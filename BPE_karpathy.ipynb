```python
text = "Perfect! Here’s a **single cell** you can drop into any notebook. It’ll ask you for a commit message and then do `add + commit + push` in one go. If you want, I can also tweak it so that it **automatically skips pushing if there’s nothing to commit**, so it won’t spam errors for empty commits. Do you want me to add."
tokens = text.encode('utf-8') #these are the raw bytes

tokens = list(map(int,tokens)) # convert to a list of integers
print('----')
print(text)
print("length:",len(text))
print('----')
print(tokens)
print("length:",len(tokens))
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids,ids[1:]):
        counts[pair]=counts.get(pair,0)+1
    return counts

stats =get_stats(tokens)
print(stats)
```
