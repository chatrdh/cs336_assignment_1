```python
def merge(ids, pair, idx):
    newids = []
    skip = False
    for i in range(len(ids)):
        if skip:
            skip = False
            continue
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            # Merge the pair into a new token
            newids.append(idx)
            skip = True  # skip the next one because it's merged
        else:
            newids.append(ids[i])
    return newids   
            
```
